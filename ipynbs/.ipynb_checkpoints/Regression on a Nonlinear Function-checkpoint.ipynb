```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression

%matplotlib inline
```

```python
N   = 500
a   = 1.0
b   = 0.0
eps = 0.5

x = np.matrix(np.random.randn(N)).T
y = np.matrix(a * x + b) + np.matrix(eps * np.random.randn(N)).T

plt.scatter(np.array(x).reshape(-1),
            np.array(y).reshape(-1), 
            s = 0.5)
```

```python
lr = LinearRegression()
lr.fit(X = x, y = y)

y_p = lr.predict(X = x)
y_err = y_p - y
err = y_err.T * y_err / len(y_err)

x_p = np.matrix(np.arange(min(x), max(x), (max(x) - min(x)) / len(x))).T

plt.scatter(np.array(x).reshape(-1),
            np.array(y).reshape(-1),
            s = 0.5)

plt.plot(np.array(x_p).reshape(-1),
         np.array(lr.predict(X = x_p)).reshape(-1), color = 'r')

plt.plot(np.array(x_p).reshape(-1),
         np.array(lr.predict(X = x_p) + np.sqrt(err)).reshape(-1), 
         color = 'r', 
         alpha = 0.2, 
         linewidth = 1.0)

plt.plot(np.array(x_p).reshape(-1),
         np.array(lr.predict(X = x_p) - np.sqrt(err)).reshape(-1), 
         color = 'r', 
         alpha = 0.2,
         linewidth = 1.0)
```

```python
plt.scatter(np.array(x).reshape(-1),
            np.array(np.tanh(y)).reshape(-1),
            s = 0.5)

plt.plot(np.array(x_p).reshape(-1),
         np.array(np.tanh(lr.predict(X = x_p))).reshape(-1), 
         color = 'r')

plt.plot(np.array(x_p).reshape(-1),
         np.array(np.tanh(lr.predict(X = x_p) + np.sqrt(err))).reshape(-1), 
         color = 'r', 
         alpha = 0.2, 
         linewidth = 1.0)

plt.plot(np.array(x_p).reshape(-1),
         np.array(np.tanh(lr.predict(X = x_p) - np.sqrt(err))).reshape(-1), 
         color = 'r', 
         alpha = 0.2,
         linewidth = 1.0)
```
