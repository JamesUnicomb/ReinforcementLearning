```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression

%matplotlib inline
```

```python
N   = 100
a   = 1.0
b   = 0.0
eps = 0.15

x = np.matrix(np.random.randn(N)).T
y = np.matrix(a * x + b) + np.matrix(eps * np.random.randn(N)).T

plt.scatter(np.array(x).reshape(-1),
            np.array(y).reshape(-1), 
            s = 0.5,
            color = 'k')
```

```python
lr = LinearRegression()
lr.fit(X = x, y = y)

y_p = lr.predict(X = x)
y_err = y_p - y
err = y_err.T * y_err / len(y_err)

x_p = np.matrix(np.arange(min(x), max(x), (max(x) - min(x)) / len(x))).T

plt.scatter(np.array(x).reshape(-1),
            np.array(y).reshape(-1),
            s = 0.5,
            color = 'k')

plt.plot(np.array(x_p).reshape(-1),
         np.array(lr.predict(X = x_p)).reshape(-1),
         color = 'r',
         linewidth = 3.0)

plt.plot(np.array(x_p).reshape(-1),
         np.array(lr.predict(X = x_p) + 2 * np.sqrt(err)).reshape(-1),
         color = 'r',
         alpha = 0.2,
         linewidth = 2.0)

plt.plot(np.array(x_p).reshape(-1),
         np.array(lr.predict(X = x_p) - 2 * np.sqrt(err)).reshape(-1), 
         color = 'r',
         alpha = 0.2,
         linewidth = 2.0)
```

```python
plt.scatter(np.array(x).reshape(-1),
            np.array(np.tanh(y)).reshape(-1),
            s = 0.5,
            color = 'k')

plt.plot(np.array(x_p).reshape(-1),
         np.array(np.tanh(lr.predict(X = x_p))).reshape(-1), 
         color = 'r',
         linewidth = 3.0)

plt.plot(np.array(x_p).reshape(-1),
         np.array(np.tanh(lr.predict(X = x_p) + 2 * np.sqrt(err))).reshape(-1), 
         color = 'r', 
         alpha = 0.2, 
         linewidth = 2.0)

plt.plot(np.array(x_p).reshape(-1),
         np.array(np.tanh(lr.predict(X = x_p) - 2 * np.sqrt(err))).reshape(-1), 
         color = 'r', 
         alpha = 0.2,
         linewidth = 2.0)
```

```python
mu = 0.0
sg = 0.1

x = np.arange(mu - 3 * sg, mu + 3 * sg, 6 * sg / 100)
y = 1 / np.sqrt(2 * np.pi * sg * sg) * np.exp( - (mu - x) * (mu - x) / (2 * sg * sg))
lny = - 1 / 2 * np.log(2 * np.pi * sg * sg) - (mu - x) * (mu - x) / (2 * sg * sg)
dlny = - (mu - x) / (sg * sg)

plt.figure(figsize = (9,3))

plt.subplot(131)
plt.plot(x,y)
plt.xlim([mu - 3.3 * sg, mu + 3.3 * sg])

plt.subplot(132)
plt.plot(x,lny)
plt.xlim([mu - 3.3 * sg, mu + 3.3 * sg])

plt.subplot(133)
plt.plot(x,dlny)
plt.xlim([mu - 3.3 * sg, mu + 3.3 * sg])
```
